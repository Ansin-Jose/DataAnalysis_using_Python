**Data Cleaning**

**Find all the null values in dataset.If there is any null values in the column, then fill it with mean values in the column.**

```python
data.isnull().sum()
```

```python
data['Cylinders'].fillna(data['Cylinders'].mean(), inplace = True)
```

```python
data.isnull().sum()
```

**Check out what are the different types of make are there in our dataset.And what is the count(occurrence) of each Make in the data?**

```python
data['Make'].value_counts()
```

**Show all the records where origin is Asia or Europe**

```python
data[(data['Origin'] == 'Asia')| (data['Origin'] == 'Europe')]
```

```python
data[data['Origin'].isin(['Asia','Europe'])]
```

**Remove all the records(rows) where is above 4000**

```python
data[~(data['Weight'] > 4000)]
```

**Increase all the values of 'MPG_City' column by 3**

```python
data.head(10)
```
